```python
class QueueList:
  def __init__(self):
    self.queue = []
  def enqueue(self, item):
    self.queue.append(item)
  def dequeue(self):
    if not self.is_empty():
      return self.queue.pop(0)
    return None
  def front(self):
    return self.queue[0] if not self.is_empty() else None
  def rear(self):
    return self.queue[-1] if not self.is_empty() else None
  def is_empty(self):
    return len(self.queue) == 0
  def size(self):
    return len(self.queue)

queue = QueueList()
queue.enqueue(70)
queue.enqueue(100)
queue.enqueue(53)
queue.enqueue(17)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("front Element:", queue.front())
print("rear Element:", queue.rear())
print("Is queue empty?", queue.is_empty())
print("Size of queue:", queue.size())
```

```python
from collections import deque

class QueueDeque:
  def __init__(self):
    self.queue = deque()
  def enqueue(self, item):
    self.queue.append(item)
  def dequeue(self):
    return self.queue.popleft() if not self.is_empty() else None
  def front(self):
    return self.queue[0] if not self.is_empty() else None
  def rear(self):
    return self.queue[-1] if not self.is_empty() else None
  def is_empty(self):
    return len(self.queue) == 0
  def size(self):
    return len(self.queue)

queue = QueueDeque()
queue.enqueue(70)
queue.enqueue(100)
queue.enqueue(53)
queue.enqueue(17)
print("Queue setelah enqueue:", queue.queue)
print("Dequeued element:", queue.dequeue())
print("front Element:", queue.front())
print("rear Element:", queue.rear())
print("Is queue empty?", queue.is_empty())
print("Size of queue:", queue.size())
```

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
class QueueLinkedList:
  def __init__(self):
    self.front = self.rear = None
  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
      self.front = self.rear = new_node
      return
    self.rear.next = new_node
    self.rear = new_node
  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    return dequeued_element
  def is_empty(self):
    return self.front is None
  def front_element(self):
    return self.front.data if self.front else None
  def rear_element(self):
    return self.rear.data if self.rear else None

queue = QueueLinkedList()
queue.enqueue(70)
queue.enqueue(100)
queue.enqueue(53)
queue.enqueue(17)
print("Dequeued element:", queue.dequeue())
print("front Element:", queue.front_element())
print("rear Element:", queue.rear_element())
print("Is queue empty?", queue.is_empty())
```

```python
class PhotocopyQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Fast Track: Pesanan {order_id} langsung diproses!")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} dengan {num_pages} halaman ditambahkan ke antrian.")

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan yang dapat diproses.")
        else:
            order_id, num_pages = self.queue.pop(0)
            if num_pages > 100:
                remaining_pages = num_pages - 100
                print(f"Pesanan {order_id} diproses 100 halaman, sisa {remaining_pages} halaman tetap dalam antrian.")
                self.queue.insert(0, (order_id, remaining_pages))
            else:
                print(f"Pesanan {order_id} selesai diproses.")

    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan pertama dalam antrian: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan terakhir dalam antrian: {self.queue[-1]}")

# Contoh Penggunaan
queue = PhotocopyQueue()
queue.enqueue("ALI23", 50)  # Pesanan normal
queue.enqueue("BRI47", 5)   # Fast Track
queue.enqueue("CAR56", 120) # Pesanan lebih dari 100 halaman
queue.front()
queue.rear()
queue.dequeue()
queue.front()
queue.size()
```

```python
class ParkingQueue:
    def __init__(self, capacity=15):
        self.queue = []
        self.capacity = capacity

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, vehicle_number, is_electric=False):
        if is_electric:
            self.queue.insert(0, vehicle_number)
            print(f"Parkir Prioritas: Kendaraan listrik {vehicle_number} mendapatkan prioritas masuk.")
        elif self.size() < self.capacity:
            self.queue.append(vehicle_number)
            print(f"Kendaraan {vehicle_number} ditambahkan ke antrian parkir.")
        else:
            print("Parkiran penuh! Harap tunggu hingga ada kendaraan yang keluar.")

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan yang keluar.")
        else:
            exited_vehicle = self.queue.pop(0)
            print(f"Kendaraan {exited_vehicle} telah keluar dari parkiran.")

    def front(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan pertama dalam antrian: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan terakhir dalam antrian: {self.queue[-1]}")

# Contoh Penggunaan
parking = ParkingQueue()
parking.enqueue("JA23")  # Kendaraan biasa
parking.enqueue("SU47", is_electric=True)  # Kendaraan listrik (prioritas)
parking.enqueue("BD56")  # Kendaraan biasa
parking.front()
parking.rear()
parking.dequeue()
parking.front()
parking.size()
```

```python
class CertificateQueue:
    def __init__(self, capacity=20):
        self.regular_queue = []
        self.achievement_queue = []
        self.capacity = capacity

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.achievement_queue) == 0

    def size(self):
        return len(self.regular_queue) + len(self.achievement_queue)

    def enqueue(self, student_id, is_achiever=False):
        if is_achiever:
            print(f"Mahasiswa berprestasi {student_id} langsung dapat mengambil sertifikat!")
        elif self.size() >= self.capacity:
            self.achievement_queue.append(student_id)
            print(f"Antrian penuh, {student_id} dimasukkan ke jalur mahasiswa berprestasi.")
        else:
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} ditambahkan ke antrian reguler.")

    def dequeue(self):
        if not self.is_empty():
            if self.achievement_queue:
                served_student = self.achievement_queue.pop(0)
            else:
                served_student = self.regular_queue.pop(0)
            print(f"Mahasiswa {served_student} telah mengambil sertifikat.")
        else:
            print("Antrian kosong, tidak ada mahasiswa yang dilayani.")

    def front(self):
        if self.achievement_queue:
            print(f"Mahasiswa pertama dalam jalur prestasi: {self.achievement_queue[0]}")
        elif self.regular_queue:
            print(f"Mahasiswa pertama dalam jalur reguler: {self.regular_queue[0]}")
        else:
            print("Antrian kosong.")

    def rear(self):
        if self.regular_queue:
            print(f"Mahasiswa terakhir dalam jalur reguler: {self.regular_queue[-1]}")
        elif self.achievement_queue:
            print(f"Mahasiswa terakhir dalam jalur prestasi: {self.achievement_queue[-1]}")
        else:
            print("Antrian kosong.")

# Contoh Penggunaan
certificate_queue = CertificateQueue()
certificate_queue.enqueue("T23")  # Mahasiswa reguler
certificate_queue.enqueue("S47", is_achiever=True)  # Mahasiswa berprestasi
certificate_queue.enqueue("B56")  # Mahasiswa reguler
certificate_queue.front()
certificate_queue.rear()
certificate_queue.dequeue()
certificate_queue.front()
certificate_queue.size()
```

```python
#latihan 4
class AntrianSertifikat:
    def __init__(self, capacity=20):
        self.regular_queue = []
        self.pencapaian = []
        self.capacity = capacity

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.pencapaian) == 0

    def size(self):
        return len(self.regular_queue) + len(self.pencapaian)

    def enqueue(self, student_id, prestasi=False):
        if prestasi:
            print(f"Mahasiswa berprestasi {student_id} langsung dapat mengambil sertifikat!")
        elif self.size() >= self.capacity:
            self.pencapaian.append(student_id)
            print(f"Antrian penuh, {student_id} dimasukkan ke jalur mahasiswa berprestasi.")
        else:
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} ditambahkan ke antrian reguler.")

    def dequeue(self):
        if not self.is_empty():
            if self.pencapaian:
                served_student = self.pencapaian.pop(0)
            else:
                served_student = self.regular_queue.pop(0)
            print(f"Mahasiswa {served_student} telah mengambil sertifikat.")
        else:
            print("Antrian kosong, tidak ada mahasiswa yang dilayani.")

    def front(self):
        if self.pencapaian:
            print(f"Mahasiswa pertama dalam jalur prestasi: {self.pencapaian[0]}")
        elif self.regular_queue:
            print(f"Mahasiswa pertama dalam jalur reguler: {self.regular_queue[0]}")
        else:
            print("Antrian kosong.")

    def rear(self):
        if self.regular_queue:
            print(f"Mahasiswa terakhir dalam jalur reguler: {self.regular_queue[-1]}")
        elif self.pencapaian:
            print(f"Mahasiswa terakhir dalam jalur prestasi: {self.pencapaian[-1]}")
        else:
            print("Antrian kosong.")

# Contoh Penggunaan
antrean = AntrianSertifikat()
antrean.enqueue("T53")
antrean.enqueue("S47", prestasi=True)
antrean.enqueue("B56")
antrean.enqueue("G20")
print("Antrean setelah enqueue:", antrean.regular_queue)
antrean.front()
antrean.rear()
antrean.dequeue()
antrean.front()
antrean.size()
```
