```python
class SatpamNode:
  def __init__(self, nama):
    self.nama = nama
    self.next = None
class SatpamList:
  def __init__(self):
    self.head = None
  def add_satpam(self, nama):
    new_node = SatpamNode(nama)
    if self.head is None:
      self.head = new_node
      self.head.next = self.head
    else:
      current = self.head
      while current.next != self.head:
        current = current.next
      current.next = new_node
      new_node.next = self.head
  def display_satpam(self):
    if self.head is None:
      print("Tidak ada satpam yang terdaftar.")
      return
    print("Daftar Satpam:")
    current = self.head
    count = 1
    while True:
      print(current.nama)
      current = current.next
      count += 1
      if current == self.head:
        break
  def Roling(self, days):
    if self.head is None:
      print("Tidak ada satpam yang terdaftar.")
      return
    jumlah_satpam = self.count_satpam()
    current = self.head
    for _ in range(days % jumlah_satpam):
        current = current.next
    return current.nama
  def count_satpam(self):
    count = 0
    current = self.head
    if current is None:

      return count
    while True:
        count += 1
        current = current.next
        if current == self.head:
            break
    return count
# main program
#user mamasukkan jumlah satpam
ListSatpam = SatpamList()
jumlah_satpam = int(input("Masukkan jumlah satpam yang ingin ditambahkan: "))
for i in range(4):
  name = input("\nMasukkan nama satpam: ")
  ListSatpam.add_satpam(name)
ListSatpam.display_satpam()
print("\nJadwal Piket Satpam selama 2 minggu: ")
for day in range(1, 15):
  satpam = ListSatpam.Roling(day - 1)
  print(f"Hari ke-{day}: {satpam}")
```
