```python
#definisi node dalam linked list
class MatkulNode:
  def __init__(self, matkul):
    self.matkul = matkul #menyimpan nama Matkul
    self.next = None #menyimpan alamat dari Matkul selanjutnya
#Definisi Linked list untuk Daftar matakuliah
class DaftarMatkul:
  def __init__(self):
    self.head = None #menyimpan alamat dari Matkul pertama

  #menambah matakuliah
  def add_matkul(self, matkul):
    new_matkul = MatkulNode(matkul)
    if not self.head:
      self.head = new_matkul #jika list kosong matkul pertama menjadi head
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next = new_matkul
  #menampilkan matakuliah
  def display_matkul(self):
    if not self.head:
      print("\n Tidak ada matakuliah hari ini:")
      return
    print("\n Daftar Matakuliah hari ini:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.matkul}")
      current = current.next
      count += 1
#Program utama
daftar_matkul = DaftarMatkul()
#meminta pengguna memasukkan jumlah matkul hari ini
jumlah_matkul = int(input("Masukkan jumlah matakuliah hari ini: "))
#meminta pengguna memasukkan nama matakuliah
for i in range(1, jumlah_matkul + 1):
  matkul = input(f"Masukkan nama matakuliah ke-{i}: ")
  daftar_matkul.add_matkul(matkul)
#menampilkan daftar matakuliah
daftar_matkul.display_matkul()
```

```python
#Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url):
    self.url = url #URL halaman web
    self.next = None #Pointer ke halaman berikutnya
    self.prev = None #pointer ke halaman sebelumnya
#Definisi Doubly Linked List
class BrowserHistory:
  def __init__(self):
    self.current = None #Pointer ke halaman yang sedang dibuka

  #Menambahkan halaman baru kedalam riwayat
  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page #jika riwayat kosong, halaman ini menjadi halaman pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page
  def show_history(self):
    if not self.current:
      print("\n Browser history is empty.")
      return
    print("\n Browser history:")
    temp = self.current
    while temp.prev:
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp = temp.next
      count += 1
  def go_back(self):
    if self.current and self.current.prev:
       self.current = self.current.prev
       print(f"\n anda maju ke: {self.current.url}")
    else:
      print("\n tidak bisa maju, ini adalah halaman pertama!")
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.url}")
    else:
      print("\ntidak bisa maju, ini adalah halaman terakhir")

#MAIN PROGRAM
history = BrowserHistory()
num_pages = int(input("\nMasukkan jumlah halaman yang di kunjungi: "))
for i in range(1, num_pages +1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)
history.show_history()

while True:
  command = input("\n Tekan 'b' untuk kembali, tekan 'f' untuk maju, atau tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat peramban.")
    break
  else:
    print("\n perintah tidak dikenal. ")
```

```python
#definisi node dalam circular linked list
class SongNode:
  def __init__(self, titel):
    self.titel = titel #menyimpan judul lagu
    self.next = None #pointer ke lagu berikurnya
#definisi circular linked list unntuk playlist
class Playlist:
  def __init__(self):
    self.current = None #lagu yang sedang di putar
    #menambahkan lagu kedalam play list
  def add_song(self, titel):
    new_song = SongNode(titel)
    if not self.current:
      self.current = new_song
      self.current.next = self.current #menghubungkan ke dirinya sendiri
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next #mencari lagu terakhir
      temp.next = new_song #menambahkan lagu baru ke akhir
      new_song.next = self.current #menghubungkan lagu baru ke lagu pertama
    #menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print("\n Playlist Kosong!.")
      return
    print("\n Playlist lagu anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.titel}")
      temp = temp.next
      count += 1
      if temp == self.current: #berhenti jika sudah kembali ke awal
        break
    #memutar lagu berikutnya dalam playlist
  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar lagu: {self.current.titel}")

#main program
#membuat objek play list
Playlist = Playlist()
#meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))
#meminta pengguna memasukkan judul lagu kedalam circular linked list
for i in range(1, num_songs + 1):
  titel = input(f"Masukkan judul lagu ke-{i}: ")
  Playlist.add_song(titel)
#menampilkan daftar lagu dalam playlist
Playlist.show_playlist()
#menu navigasi pengguna
while True:
  command = input("\n Tekan 'n' untuk memutar lagu berikutnya, atau tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()
  if command == 'n':
    Playlist.next_song()
  elif command == 'x':
    print("\n Keluar dari playlist.")
    break
  else:
    print("\n perintah tidak dikenal.")

```
