```python
class Stack:
  def __init__(self):
    self.stack = []
  def push(self, item):
    self.stack.append(item)
  def pop(self):
    if len(self.stack) == 0:
      return "stack kosong"
    return self.stack.pop()
  def peek(self):
    if len(self.stack) == 0:
      return "Stack kosong"
    return self.stack[-1]


s = Stack()
s.push (5)
s.push (10)
s.push (15)
s.push (20)
print(s.pop())
print(s.stack)
print(s.peek())#melihat item terakhir tanpa menghapusnya
print(s.stack)
```

```python
class TextEditor:
  def __init__(self):
    self.history = []
  def write(self, text):
    self.history.append(text)
  def undo(self):
    if self.history:
      return f"undo: Menghapus '{self.history.pop()}'"
    return "Tidak ada yang bisa di undo"

editor = TextEditor()
editor.write("halo,ini teks pertama.")
editor.write('ini teks kedua.')
editor.write('ini teks ketiga.')

print(editor.undo())
print(editor.undo())
print(editor.undo())

```

```python
#latihan 1
class Kalkulator:
  def __init__(self):
    self.operasi_matematika = []
  def write(self, angka):
    self.operasi_matematika.append(angka)
  def undo(self):
    if self.operasi_matematika:
      return f"undo: Menghapus '{self.operasi_matematika.pop()}'"
    return "Tidak ada yang bisa di undo"

num_angka = int(input("\n Masukan jumlah operasi matemtaika: "))
kalkulator = Kalkulator()
for i in range(num_angka):
  hasil_operasi = int(input('\n Masukkan hasil operasi matematika: '))
  kalkulator.write(hasil_operasi)
print(f"history operasi matematika {kalkulator.operasi_matematika}")
button=input("ketik 'f' untuk undo: ")
if button == 'f':
  print(kalkulator.undo())
else:
  print("tombol undo tidak valid")
```

```python
from collections import deque

class BrowserHistory:
  def __init__(self):
    self.history = deque()
  def visit(self, url):
    self.history.append(url)
    print(f"berhasil mengunjungi: '{url}'")
  def back(self):
    if self.history:
      print(f"kembali ke: {self.history.pop()}")
    else:
      print('Tidak ada riwayat!')

browser = BrowserHistory()
browser.visit("google.com")
browser.visit("youtube.com")
browser.visit("github.com")

browser.back()
browser.back()

```

```python
from queue import LifoQueue

class BackupSystem:
  def __init__(self):
    self.backups = LifoQueue()
  def backup(self, version):
    self.backups.put(version)
  def latest_backup(self):
    if not self.backups.empty():
      return f"Versi terbaru:{self.backups.queue[-1]}"
    return "Tidak ada cadangan!"

backup_system = BackupSystem()
backup_system.backup("Backup_1")
backup_system.backup("Backup_2")
backup_system.backup("Backup_3")

print(backup_system.latest_backup())
```

```python
class CustomerService:
  def __init__(self):
    self.tickets = []
  def add_tickets(self, ticket):
    self.tickets.append(ticket)
  def process_ticket(self):
    if not self.is_empty():
      return f"Memproses tiket: {self.tickets.pop()}"
    return "Tidak ada tiket yang harus diproses."
  def is_empty(self):
    return len(self.tickets) == 0

service = CustomerService()
service.add_tickets("Tiket 001")
service.add_tickets("Tiket 002")

print(service.process_ticket())
print(service.is_empty())
```

```python
#Latihan 2
class NavigasiHistory:
  def __init__(self):
    self.history = []
  def visit(self, fitur):
    self.history.append(fitur)
    print(f"berhasil mengunjungi: '{fitur}'")
  def back(self):
    if len(self.history) > 1:
      self.history.pop()
      print(f"kembali ke: {self.history[-1]}")
    else:
      print('Tidak ada riwayat!')
  def tampilkan_history(self):
        print("Riwayat fitur yang dikunjungi:")
        for fitur in reversed(self.history):
            print(f"- {fitur}")

navigasi = NavigasiHistory()

navigasi.visit("F23")
navigasi.visit("J24")
navigasi.visit("J25")
navigasi.tampilkan_history()
navigasi.back()
navigasi.back()
navigasi.back()
```

```python
#Tugas
class TowerOfHanoi:
  def __init__(self, n):
    self.n = n
    self.tower = {
        'A': list(range(n, 0, -1)),
        'B': [],
        'C': []
    }
    self.langkah = 0
  def pindah_cakram(self, asal, tujuan):
    if self.tower[asal] and (not self.tower[tujuan] or self.tower[asal][-1] < self.tower[tujuan][-1]):
      cakram = self.tower[asal].pop()
      self.tower[tujuan].append(cakram)
      self.langkah += 1
      print(f"Memindahkan cakram {cakram} dari {asal} ke {tujuan}")
    else:
      print("Pindah cakram tidak valid")
  def tampilkan_tiang(self):
    for tower, cakram in self.tower.items():
      print(f"{tower}: {cakram}")
  def mainkan_hanoi(self, n, asal, tujuan, bantu):
    if n == 1:
      self.pindah_cakram(asal, tujuan)
    else:
      self.mainkan_hanoi(n-1, asal, bantu, tujuan)
      self.pindah_cakram(asal, tujuan)
      self.mainkan_hanoi(n-1, bantu, tujuan, asal)

maks_langkah = 53
game = TowerOfHanoi(3)
print("keadaan awal:")
game.tampilkan_tiang()
if game.langkah < maks_langkah:
  game.mainkan_hanoi(3, 'A', 'C', 'B')
print("keadaan akhir:")
game.tampilkan_tiang()
```

```python
class TowerOfHanoi:
    def __init__(self, n):
        self.n = n  # Jumlah cakram
        self.tiang = {
            'A': list(range(n, 0, -1)),  # Tiang awal dengan cakram terbesar di bawah
            'B': [],  # Tiang kosong
            'C': []   # Tiang kosong
        }
        self.langkah = 0

    def pindahkan_cakram(self, asal, tujuan):
        if self.tiang[asal] and (not self.tiang[tujuan] or self.tiang[asal][-1] < self.tiang[tujuan][-1]):
            cakram = self.tiang[asal].pop()
            self.tiang[tujuan].append(cakram)
            self.langkah += 1
            print(f"Memindahkan cakram {cakram} dari {asal} ke {tujuan}")
        else:
            print("Pindah tidak valid!")

    def tampilkan_tiang(self):
        print("\nKeadaan Tiang:")
        max_height = max(len(self.tiang['A']), len(self.tiang['B']), len(self.tiang['C']))
        for i in range(max_height, 0, -1):
            for tiang in ['A', 'B', 'C']:
                if len(self.tiang[tiang]) >= i:
                    print(f" {self.tiang[tiang][i-1]} ", end="  ")
                else:
                    print(" | ", end="  ")
            print()
        print(" A    B    C ")

    def mainkan_hanoi(self, n, asal, tujuan, bantu):
        if n == 1:
            self.pindahkan_cakram(asal, tujuan)
            self.tampilkan_tiang()
        else:
            self.mainkan_hanoi(n-1, asal, bantu, tujuan)
            self.pindahkan_cakram(asal, tujuan)
            self.tampilkan_tiang()
            self.mainkan_hanoi(n-1, bantu, tujuan, asal)

# Gantilah dua digit terakhir NPM sebagai batas langkah
maks_langkah = 23  # Misalkan NPM berakhir dengan 23

game = TowerOfHanoi(3)
print("Keadaan awal:")
game.tampilkan_tiang()

# Menjalankan permainan otomatis jika diizinkan
if game.langkah < maks_langkah:
    game.mainkan_hanoi(3, 'A', 'C', 'B')

print("Keadaan akhir:")
game.tampilkan_tiang()
```
