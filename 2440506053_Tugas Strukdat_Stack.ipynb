```python
class TowerOfHanoi:
    def __init__(self, n):
        self.n = n  # Jumlah cakram
        self.tiang = {
            'A': list(range(n, 0, -1)),  # Tiang awal dengan cakram terbesar di bawah
            'B': [],  # Tiang kosong
            'C': []   # Tiang kosong
        }
        self.langkah = 0

    def pindahkan_cakram(self, asal, tujuan):
        if self.tiang[asal] and (not self.tiang[tujuan] or self.tiang[asal][-1] < self.tiang[tujuan][-1]):
            cakram = self.tiang[asal].pop()
            self.tiang[tujuan].append(cakram)
            self.langkah += 1
            print(f"Memindahkan cakram {cakram} dari {asal} ke {tujuan}")
        else:
            print("Pindah tidak valid!")

    def tampilkan_tiang(self):
        print("\nKeadaan Tiang:")
        max_height = max(len(self.tiang['A']), len(self.tiang['B']), len(self.tiang['C']))
        for i in range(max_height, 0, -1):
            for tiang in ['A', 'B', 'C']:
                if len(self.tiang[tiang]) >= i:
                    print(f" {self.tiang[tiang][i-1]} ", end="  ")
                else:
                    print(" | ", end="  ")
            print()
        print(" A    B    C ")

    def mainkan_hanoi(self, n, asal, tujuan, bantu):
        if n == 1:
            self.pindahkan_cakram(asal, tujuan)
            self.tampilkan_tiang()
        else:
            self.mainkan_hanoi(n-1, asal, bantu, tujuan)
            self.pindahkan_cakram(asal, tujuan)
            self.tampilkan_tiang()
            self.mainkan_hanoi(n-1, bantu, tujuan, asal)


maks_langkah = 53
game = TowerOfHanoi(3)
print("Keadaan awal:")
game.tampilkan_tiang()

# Menjalankan permainan otomatis jika diizinkan
if game.langkah < maks_langkah:
    game.mainkan_hanoi(3, 'A', 'C', 'B')

print("Keadaan akhir:")
game.tampilkan_tiang()
```
