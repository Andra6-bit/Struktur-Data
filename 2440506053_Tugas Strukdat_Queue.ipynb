```python
#latihan 2
class AntrianFotokopi:
    def __init__(self):
        self.queue = []
    def enqueue(self, nama_pemesan, jumlah_halaman):
        if jumlah_halaman < 10:
            print(f"Fast Track: Pesanan {nama_pemesan} langsung diproses!")
        else:
            self.queue.append((nama_pemesan, jumlah_halaman))
            print(f"Pesanan {nama_pemesan} dengan {jumlah_halaman} halaman ditambahkan ke antrian.")
    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong, tidak ada pesanan yang dapat diproses.")
        else:
            nama_pemesan, jumlah_halaman = self.queue.pop(0)
            if jumlah_halaman > 100:
                sisa_halaman = jumlah_halaman - 100
                print(f"Pesanan {nama_pemesan} diproses 100 halaman, sisa {sisa_halaman} halaman tetap dalam antrian.")
                self.queue.insert(0, (nama_pemesan, sisa_halaman))
            else:
                print(f"Pesanan {nama_pemesan} selesai diproses.")
    def front(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan pertama dalam antrian: {self.queue[0]}")
    def rear(self):
        if self.is_empty():
            print("Antrian kosong.")
        else:
            print(f"Pesanan terakhir dalam antrian: {self.queue[-1]}")

    def is_empty(self):
        return len(self.queue) == 0
    def size(self):
        return len(self.queue)

antrean = AntrianFotokopi()
antrean.enqueue('AND53', 50)
antrean.enqueue('BRI47', 5)
antrean.enqueue('CAR56', 120)
antrean.enqueue('RIO', 30)
print("Antrean setelah enqueue:", antrean.queue)
antrean.front()
antrean.rear()
antrean.dequeue()
antrean.front()
antrean.size()
```

```python
#Latihan 3
class AntrianParkir:
    def __init__(self, capacity=15):
        self.queue = []
        self.capacity = capacity

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, nomor_plat, ramah_lingkungan=False):
        if ramah_lingkungan:
            self.queue.insert(0, nomor_plat)
            print(f"Parkir Prioritas: Kendaraan listrik {nomor_plat} mendapatkan prioritas masuk.")
        elif self.size() < self.capacity:
            self.queue.append(nomor_plat)
            print(f"Kendaraan {nomor_plat} ditambahkan ke antrian parkir.")
        else:
            print("Parkiran penuh! Harap tunggu hingga ada kendaraan yang keluar.")

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong, tidak ada kendaraan yang keluar.")
        else:
            exited_vehicle = self.queue.pop(0)
            print(f"Kendaraan {exited_vehicle} telah keluar dari parkiran.")

    def front(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan pertama dalam antrian: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("Parkiran kosong.")
        else:
            print(f"Kendaraan terakhir dalam antrian: {self.queue[-1]}")

# Contoh Penggunaan
parking = AntrianParkir()
parking.enqueue("SL53")
parking.enqueue("SU47", ramah_lingkungan=True)
parking.enqueue("BD56")
parking.enqueue("LZ06")
print("Antrean setelah enqueue:", parking.queue)
parking.front()
parking.rear()
parking.dequeue()
parking.front()
parking.size()
```
